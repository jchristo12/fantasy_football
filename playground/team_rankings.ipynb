```python
def team_rank(data, stat_col):
    #group and perform the ranking
    group = data.groupby(by=['seas', 'wk'], as_index=False, sort=False)
    rank = group[stat_col].rank(method='min', ascending=True)
    #add a suffix to the data
    rank.columns = [stat_col + '_rank']
    
    return rank
```

#### Make sure the functions are working

```python
py_rank = team_rank(df, 'py')
```
